# RefineNet
Here we construct a new GAN type that builds images through succesive refinement steps. The GAN utilizes a modified Haar basis to translate an image (or noise) into another image by starting with coarse features and refining the image by progressively adding detail at each step. Critical to the construction is that compositional information at previous steps is preserved in successive steps -- each successive step retains the composition, but refines only the details.

## The modified Haar decomposition
Here we employ a modified Haar decomposition of an image. This process will be reversed in the GAN to create an image. However, the input image will be deconstructed in using this transform.

How much information do we need to add to an image when doubling its size? Consider a $2^n\times 2^n$ image, that we wish to double to $2^{n+1}\times 2^{n+1}$. How many additional pixels are added? The answer is:
$$
\begin{eqnarray}
2^{n+1}\times 2^{n+1} - 2^{n}\times 2^{n} &=& 2^{2(n+1)}-2^{2n} \\
&=& (2^2-1)2^{2n} \\
&=& 3\cdot 2^{n}\times 2^{n}\,.
\end{eqnarray}
$$
The easiest way to vizualize this is by taking the original image to be the top left corner of the doubled-image. There are three remaining corners that need to be filled in: top-right, bottom-left, and bottom-right. Each of these corners represents $2^n\times 2^n$ additional pixels. Hence, we need to fill in $3\cdot 2^{n}\times 2^{n}$ additional pixels.

<Insert Image>

BUT, there is another way to fill in the image. Suppose we wanted to use the original image as the <i>composition</i> of the doubled image, but fill in the details by making use of the enhanced resolution. To accomplish this we could simply double the original image by upscaling each pixel into a $2x2$ block. We want to preserve the <i>average</i> value of each $2\times 2$ pixel block (thereby preserving the composition of the original image), but we want to fill in some extra detail in each block. 

<Insert Image>

There are 4 pixels in each block, so we need at least 4 new numbers. But, there is one constraint: that the average value of each pixel block is equal to the original pixel value. Thus, for each pixel block, in theory we only need to introduce 3 new numbers. Since there are $2^n\times 2^n$ pixel blocks to fill in, we need $3\cdot 2^n\times 2^n$ new parameters -- the same value as calculated above.

How can we solve the constraint and accomplish this with just three numbers? We can use a modified version of Haar wavelets. Wavelets are a discrete, orthonormal basis on L2-space that are formed from scaling and translating a "mother" wavelet. Haar wavelets were historically the first wavelet basis ever discovered (by Haar). Here we use a modified Haar basis that is more useful in this scenario. 

The modified-Haar basis starts with 4 $2\times 2$ kernels representing the 4 pieces of information need to construct a $2\times 2$ pixel block. They correspond to the average value of the block, vertical lines, horizontal lines, and diagonal lines:
$$
k_a = \left[\begin{matrix} 1 & 1 \\ 1 & 1\end{matrix}\right] \quad \quad
k_v = \left[\begin{matrix} 1 & 1 \\ -1 & -1\end{matrix}\right] \quad \quad
k_h = \left[\begin{matrix} -1 & 1 \\ -1 & 1\end{matrix}\right] \quad \quad
k_d = \left[\begin{matrix} 1 & -1 \\ -1 & 1\end{matrix}\right]\,.
$$
We can define an inner product by 
$$
\langle u, v \rangle = \frac{1}{4} u_{ij}v^{ij} =\text{mean}\left(u * v \right)
$$
where $*$ represents component-wise multiplication. Under this inner product, the kernels, here indexed by $\{a, v, h, d\}=\{0,1,2,3\}$, are orthonormal:
$$
\langle k_{I} k_{J} \rangle = \delta_{IJ}\,.
$$
An arbitrary $2\times 2$ matrix can be represented as
$$
M = \alpha_a \left[\begin{matrix} 1 & 1 \\ 1 & 1\end{matrix}\right]
+ \alpha_v \left[\begin{matrix} 1 & 1 \\ -1 & -1\end{matrix}\right]
+ \alpha_h \left[\begin{matrix} -1 & 1 \\ -1 & 1\end{matrix}\right]
+ \alpha_d \left[\begin{matrix} 1 & -1 \\ -1 & 1\end{matrix}\right]\,.
$$
Since the $k_v$, $k_h$, and $k_d$ basis elements average to zero, the coefficient $\alpha_a$ is the average value of the $2\times 2$ pixel block. Recall in the construction above where we were seeking 3 numbers that would refine the details of a $2\times 2$ pixel block without changing its average value. The triplet $\{\alpha_v, \alpha_h, \alpha_d\}$ are exactly the three numbers we need. 

As with wavelets, these kernels can be scaled up and translated to form a complete orthonormal basis for $2^N\times 2^N$ discrete space. To do this we employ the Kronecker product. For the $2\times 2$ case the Kronecker product of two matrices is given by
$$
A \otimes B = \left[ \begin{matrix} A_{11} [B] & A_{12}[B] \\ A_{21} [B] & A_{22}[B] \end{matrix}\right]
$$
We can use the Kronecker product to scale up each matrix. Applying the Kronecker product of the vertical, horizontal, and diagonal kernels with $k_a$ times we have:
$$
k^n_I = k_I \otimes^{(n-1)} \left[\begin{matrix} 1 & 1 \\ 1 & 1\end{matrix}\right]\,.
$$
The resulting matrix, $k^{n}_I$, is $2^{n}\times 2^{n}$ dimensional, and we denote the original kernel by $k^1_I$. To form a basis on $2^N\times 2^N$ discrete space, we embed these matrices into a $2^N \times 2^N$ dimensional matrix. To do this we need to specify the position where they are embedded, so we introduce two new numbers, $p$ and $q$ the basis elements by $k^{n,p,q}_I$. The position is determined as follows. The element $k^{n, 0, 0}_I$ denotes a kernel of size $2^{n}\times 2^{n}$ positioned in the upper left corner of a matrix with all other terms equal to zero. $k^{n, q, p}$ shifts the kernel $2^{n}p$ units down and $2^{n}q$ units to the right. If the space is bounded to size $2^N\times 2^N$ the integer labels have restricted ranges:
$$
1 \leq n \leq N \\
0 \leq p,q \leq 2^{N-n}-1\,.
$$

The constructed basis elements are all orthogonal with respect to the inner product $\langle U, V \rangle = mean(U*V)$. They are not normalized as constructed, but it will be convenient to work with the non-normalized basis. 

As a sanity check, we check that we have the right number of basis elements. Here and below, we will make use of the identity 
$$
1+\sum_{n=0}^{N-1} (b-1)b^n = b^N\,.
$$
for any integer $b$ greater than zero. This can be understood of a generalization of the identity $999..999 +1 = 1000..000$ for basis $b$. At each level $n$ we have $3\cdot 2^{N-n}\times 2^{N-n}$ independent basis elements (the factor of $3$ corresponds to the three kernal types). So we just need to sum over all $n$-levels:
$$
\begin{eqnarray}
\sum_{n=1}^N 3\cdot 2^{N-n}\times 2^{N-n} &=& \sum_{n=1}^N 3\cdot 2^{2(N-n)} \\
&=& \sum_{n=1}^N 3\cdot 4^{N-n} \\
&=& \sum_{q=0}^{N-1} 3\cdot 4^q \quad \text{where} \ q \equiv N-n \\
&=& 4^N-1 \quad \text{from identity above}\\
&=& 2^N\times 2^N -1\,.
\end{eqnarray} 
$$
It appears that we are missing exactly one degree of freedom. However, the missing degree of freedom is easy to track down -- it is precisely the mean of the $2^N\times 2^N$ matrix.

In conclusion, we've constructed an orthogonal modified Haar basis that together with the total mean forms a complete basis for $2^N\times 2^N$ discrete space. 

## Implementing the Modified Haar Decomposition
The construction above gives the theoretical basis for the modified Haar decomposition that we will perform here. In practice, rather than working directly with the modified Haar basis, we will implement the decomposition using convolutions and the recomposition using successive Kronecker products. This is convenient because 2d convolutions are simple to implement in PyTorch and they are optimized for neural network training.

We first construct a list consisting of the modified Haar basis functions $k^{n}_k$. The convolution will take care of the translations, so we just need to untranslated functions. As discussed above, these can be constructed through successive applications of the Kronecker product. It will be convenient to progress from coarse grained features to fine grained features, so we will start with the largest features first. We first create a class called "Haar" that contains the lists of kernels and methods for translating back an forth between the modified Haar basis and the standard tensor representation of an image. 

```python
class Haar: 
    def __init__(self, N:int=5): 
        super().__init__()
        self.N = N
        self.base_kernels, self.all_kernels = self.build_kernel_lists(N)
    
    def build_kernel_lists(self, N:int=5):
        k_a = torch.ones(2,2)
        k_v = torch.tensor([[-1.0,1.0],[-1.0,1.0]])
        k_h = torch.tensor([[1.0,1.0],[-1.0,-1.0]])
        k_d = torch.tensor([[1.0,-1.0],[-1.0,1.0]])
        
        base_kernels = [k_a, k_v, k_h, k_d]
        all_kernels = [[k_v,k_h,k_d]]
        
        for n in range(1, N):
            # Takes the kernels at the end of the list from the last step and doubles their resolution
            kv = torch.kron(all_kernels[-1][0],k_a)
            kh = torch.kron(all_kernels[-1][1],k_a)
            kd = torch.kron(all_kernels[-1][2],k_a)
            
            all_kernels.append([kv,kh,kd]) 
            
        all_kernels.reverse()
            
        return base_kernels, all_kernels
    
    def tensorToHaar(self,x):
        # To use the convolution we need 
        # 1) x to have the shape [batches, color_channels, 2**N,2**N]
        # 2) filters to have the shape [color_channels, 1, k,k] where k is the linear dimension of the kernel
        # 3) F.conv2d to take {x,filters} as input and set groups=color_channels
        # This will ensure that each color channel gets the same filter applied to it
        
        # The output is as follows:
        # index 0: A list consisting of one element -- a tensor of shape [batches, color_channels, 1, 1]
        # index 1: A list consisting of three elements -- three tensors of shape [batches, color_channels, 1, 1]
        # index>1: A list consisting of three elements -- three tensors of shape [batches, color_channels, 2^{n-1}, 2^{n-1}]

        batches = x.shape[0]
        color_channels = x.shape[1]
        
        # The first coefficient in the list
        haar_coefficients = [[x.mean(dim=[2,3], keepdim=True)]]
        print("first haar: ", haar_coefficients[0][0].shape)

        for index, kernels in enumerate(self.all_kernels):
            # Skip the first kernel as we treated this case separately above
            # if index == 0:
            #     continue
            
            k = kernels[0].shape[-1]
            triplet = []
            
            # First reshape all three filters to [1,1,k,k] 
            # then repeat to yield shape [color_channels, 1, k,k]
            # then perform the convolution setting the groups value to the color_channels value so color channels remain independent
            for i in range(len(kernels)): 
                kernels[i] = kernels[i].unsqueeze(0).unsqueeze(0)
                kernels[i] = kernels[i].repeat(color_channels,1,1,1)
                out = F.conv2d(x, kernels[i], groups=color_channels, stride=k, padding=0)
                triplet.append(out)

            haar_coefficients.append(triplet)

        print("**first haar: ", haar_coefficients[0][0].shape)
        
        return haar_coefficients
    
    
    def haarToTensor(self, haar, level=-1, returnList=True):
        if level==-1:
            N = len(h)
        else:
            N = level
            
        tensorList = []
        
        batches = haar[0][0]
        color_channels = haar[0][1]
        
        # First reshape the kernels to [1,1,2**n,2**n]
        ka = haar_module.base_kernels[0].unsqueeze(0).unsqueeze(0)
        kv = haar_module.base_kernels[1].unsqueeze(0).unsqueeze(0)
        kh = haar_module.base_kernels[2].unsqueeze(0).unsqueeze(0)
        kd = haar_module.base_kernels[3].unsqueeze(0).unsqueeze(0)

        for i, h in enumerate(haar):
            if i == 0:
                # Start with the image as a single pixel
                tensor = h[0][0]

                delta_image = torch.zeros_like(tensor)
                raw_haar = tensor
                raw_haar = raw_haar.unsqueeze(0)
            else:
                tensor = torch.kron(tensor, ka)
                delta_v = torch.kron(h[0], kv)/(2**(2*(N-i+1)))
                delta_h = torch.kron(h[1], kh)/(2**(2*(N-i+1)))
                delta_d = torch.kron(h[2], kd)/(2**(2*(N-i+1)))

                raw_haar = h[0]+h[1]+h[2]

                delta_image = delta_v + delta_h + delta_d

                tensor += delta_image
        
            
```

```python
def generate_noise(N=5, a=1, color_channels=3, device="cpu"):
    """
    Generates square tensor of scale invariant noise of size 2^N*2^N. 
    The scale factor, a, determines the noise character.
    a = 0 : White Noise
    a = 1 : Pink Noise
    a = 2 : Brownian Noise
    """
    
    # x = torch.randn(1, color_channels, 2**N, 2**N, device=device)
    # fft = torch.fft.fft2(x)
    # fft = torch.fft.fftshift(fft)
    
    xr = torch.randn(1, color_channels, 2**N, 2**N, device=device)
    xi = torch.randn(1, color_channels, 2**N, 2**N, device=device)
    fft = torch.complex(xr, xi)

    dist = torch.zeros_like(fft, device=device).real
    
    # Calculate distance. Uses an offeset of .5 because pixels don't sit directly at (0,0)
    # This offset also avoids divide by zero at (0,0)
    for i in range(dist.shape[-2]):
        for j in range(dist.shape[-1]):
            i0 = dist.shape[-2] // 2
            j0 = dist.shape[-1] // 2
            dist[:,:,i,j] = ((i+.5-i0)/2**N)**2 + ((j+.5-j0)/2**N)**2
    
    fft = fft/((dist)**a)
    fft = torch.fft.ifftshift(fft)
    x = torch.fft.ifft2(fft).real
    
    # Normalizes the output so std is always close to 1 no matter what resolution
    # x = x/(4*2**N)
    x = x/10
    
    
    # Squeeze the output to be between zero and one
    x = (x+1)/2
    print("max {} | min {} | std {}".format(x.max(), x.min(), x.std()))
    return x
```

```python
def hue_compliment(x):
    pos = .5*(torch.abs(x)+x)
    neg = .5*(torch.abs(x)-x)
    
    value = torch.mean(neg, dim=1)
    
    white = torch.ones_like(neg)
    neg = white-neg
    neg = neg/torch.mean(neg, dim=1) * 3*value
    
    return pos + neg

x = torch.zeros(1, 3, 2, 2)
x[0, :, 0, 0] = torch.tensor([1, 0, 0])
x[0, :, 0, 1] = torch.tensor([0, .5, 0])
#x[0, :, 1, 0] = -torch.tensor([0, 0, .5])
x[0, :, 1, 1] = -torch.tensor([.5, .2, .1])

y = hue_compliment(x)

plt.imshow(torch.movedim(torch.clamp(y.squeeze(0),0,1), 0, 2))
plt.show()
```
